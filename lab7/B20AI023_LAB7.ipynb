#<h2>Question 3
<h4>Consider theConcrete Data Yeh data set. Write code for best fitting quadratic polynomial using gradient descent method.
Choose stopping condition f(x) < 0.01.</h4>

```python
data2 = pd.read_csv("Concrete_Data_Yeh.csv")
data2.head()
```

```python
x_data = data2.iloc[:,:-1].values
y_data = data2.iloc[:,-1].values
```

```python
def find_mult(var,x_data):
  val = var[len(var)-1]
  for i in range(8,0,-1):
    val += pow(x_data[len(x_data)-i],i) * var[len(x_data)-i]
  return val
#function to find the value of f(beta)
def func2(x_data,y_data,var):
  f = 0.0
  for i in range(len(x_data)):
    f += pow(find_mult(var,x_data[i]) - y_data[i],2)
  f = f / (2 * len(x_data))
  return f
```

```python
#function to find the value of gradient of f(beta)
def grad_f2(x_data,y_data,var):
  n = len(x_data)
  grad = np.array([0]*9,dtype=float)

  for j in range(8):
    for i in range(n):
      grad[j] += (8 - j) * pow(x_data[i][j],(8 - j -1)) \
      * (find_mult(var,x_data[i]) - y_data[i])
    grad[j] = grad[j] / n

  for i in range(n):
    grad[8] += (find_mult(var,x_data[i]) - y_data[i])
  grad[8] = grad[8] / n
  return grad
```

```python
#initializing the initial values and constraints
var = np.array([0]*9,dtype=float)
beta_1 = 1e-4
beta_2 = 0.9
r= 0.5
epsilon = 0.01
iter = 0

f0 = func2(x_data,y_data,var)
g0 = grad_f2(x_data,y_data,var)

# Minimizing the loss function using gradient descent
while(np.linalg.norm(g0) > epsilon and iter < 1000):
  d0, alpha = -1*g0, 1
  f1 = func2(x_data,y_data,var + (alpha * d0))
  g1 = grad_f2(x_data,y_data,var + (alpha * d0))

  # finding optimal value of alpha
  while (f1 > f0 + (alpha * beta_1 * g0.T@d0) or 
         np.dot(g1.T, d0) < (beta_2 * np.dot(g0.T, d0))):
    alpha = alpha * r
    f1 = func2(x_data,y_data,var + alpha * d0)
    g1 = grad_f2(x_data,y_data,var + alpha * d0)

  iter = iter +1
  var += alpha * d0
  f0 = f1
  g0 = g1
```

```python
print(var)
print("Function value --> ", func2(x_data,y_data,var))
print("Number of iterations --> ",iter)
```

<h2>Question 4
<h4>Write code for stochastic gradient for the above problem with 20 random points in every iterations.

```python
#initializing the initial values and constraints
var1 = np.array([0]*9,dtype=float)
beta_1 = 1e-4
beta_2 = 0.9
r= 0.5
epsilon = 0.01
iter = 0

f0 = func2(x_data,y_data,var1)

# Minimizing the loss function using stochastic gradient descent
while(iter < 100):
  # selecting random 10 data points to calculate gradient
  temp_data = data2.sample(frac = 0.019)
  x_data2 = temp_data.iloc[:,:-1].values
  y_data2 = temp_data.iloc[:,-1].values
  g0 = grad_f2(x_data2,y_data2,var1)

  d0 = -1*g0
  alpha = 1/(iter + 1)
  f1 = func2(x_data,y_data,var1 + (alpha * d0))
  if(f1 < f0):
    var1 += alpha * d0
    f0 = f1
  iter = iter+1
```

```python
print(var)
print("Function value --> ", func2(x_data,y_data,var1))
print("Number of iterations --> ",iter)
```
